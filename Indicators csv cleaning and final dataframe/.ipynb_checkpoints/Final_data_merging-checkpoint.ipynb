```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
#make a dataframe from all the indicators data
indicators_df = pd.read_csv("indicators_clean.csv")
indicators_df.head()
```

```python
#make a dataframe from the PM2.5 data (and clean it slightly)
pm25_df = pd.read_csv("../API code and csv file/PM25_data.csv")
del pm25_df['Unnamed: 0'] #remove unecessary column)
pm25_df = pm25_df.rename(columns={'Country:': 'Country Code'}) #rename column to facilitate merging
pm25_df.head()
```

```python
final_data = indicators_df.merge(pm25_df, on = 'Country Code') #merge two dataframes
```

```python
#rename columns for better comprehension
final_data = final_data.rename(columns={'City Average:': 'City Average PM2.5','Total Average:': 'Total Average PM2.5' }) 
```

```python
final_data.head() #take a look
```

```python
final_data.to_csv('final_dataframe.csv')
```
