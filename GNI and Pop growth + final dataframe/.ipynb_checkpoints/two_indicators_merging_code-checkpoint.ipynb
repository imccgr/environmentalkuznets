```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
#make a dataframe from all the indicators data
indicators_df = pd.read_csv("two_indicators.csv")
indicators_df = indicators_df.drop('Unnamed: 0', axis=1)
indicators_df.head()
```

```python
#make a dataframe from the PM2.5 data (and clean it slightly)
pm25_df = pd.read_csv("../API code and csv file/PM25_data.csv")
pm25_df = pm25_df.drop('Unnamed: 0', axis=1) #remove unecessary column)
pm25_df = pm25_df.rename(columns={'Country:': 'Country Code'}) #rename column to facilitate merging
pm25_df.head()
```

```python
final_df = indicators_df.merge(pm25_df, on = 'Country Code') #merge two dataframes
final_df.head()
```

```python
final_df.to_csv('final_indicators.csv')
```
