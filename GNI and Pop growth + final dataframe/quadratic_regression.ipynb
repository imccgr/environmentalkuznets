```python
features = ['Mean-adjusted urban population ratio','Mean-adjusted urban population ratio squared','Log-scale mean adjusted population','Log scale mean adjusted population squared','Log scale mean-adjusted GNI/capita','Log scale mean-adjusted GNI/capita squared','Log scale mean-adjusted life expectancy','Log scale mean-adjusted life expectancy squared','Female labour force ratio, mean-adjusted','Female labour force ratio, mean-adjusted squared','Population growth, mean-adjusted','Population growth, mean-adjusted squared']

X = dataset.loc[:,features].values
y = dataset.loc[:,'PM2.5 concentration by country, 2010']
```
