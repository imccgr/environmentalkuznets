```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
gni_df = pd.read_csv("gni_data.csv")
#delete useless columns
gni_df.drop(['Country Name','Series Code', 'Series Name'], inplace=True, axis=1)
gni_df = gni_df.rename(columns={'2010 [YR2010]': 'GNI 2010','2011 [YR2011]': 'GNI 2011','2012 [YR2012]': 'GNI 2012','2013 [YR2013]': 'GNI 2013','2014 [YR2014]': 'GNI 2014','2015 [YR2015]': 'GNI 2015','2016 [YR2016]': 'GNI 2016','2017 [YR2017]': 'GNI 2017','2018 [YR2018]': 'GNI 2018','2019 [YR2019]': 'GNI 2019'}) 
gni_df = gni_df.dropna() #drop the NAN which show missing data as it creates a bug when doing pivot()
gni_df.head()
```

```python
pop_growth_df = pd.read_csv("population_growth_data.csv")
#delete useless columns
pop_growth_df.drop(['Country Name','Series Code', 'Series Name'], inplace=True, axis=1)
pop_growth_df = pop_growth_df.rename(columns={'2010 [YR2010]': 'POP GROWTH 2010','2011 [YR2011]': 'POP GROWTH 2011','2012 [YR2012]': 'POP GROWTH 2012','2013 [YR2013]': 'POP GROWTH 2013','2014 [YR2014]': 'POP GROWTH 2014','2015 [YR2015]': 'POP GROWTH 2015','2016 [YR2016]': 'POP GROWTH 2016','2017 [YR2017]': 'POP GROWTH 2017','2018 [YR2018]': 'POP GROWTH 2018','2019 [YR2019]': 'POP GROWTH 2019'})
pop_growth_df = pop_growth_df.dropna() #drop the NAN which show missing data as it creates a bug when doing pivot()
pop_growth_df.head()
```

```python
two_indicators_df = gni_df.merge(pop_growth_df, on = 'Country Code') #merge two dataframes
two_indicators_df.head()
```

```python
two_indicators_df.to_csv('two_indicators.csv')
```
