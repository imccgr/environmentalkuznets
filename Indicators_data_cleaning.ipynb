```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
#import data and take a look
indicators_df = pd.read_excel("indicators_data.xlsx")
indicators_df.head()
```

```python
#delete useless columns - I know this is not the most efficient way, I will make it more efficient later - for now, this works.
del indicators_df['Country Name']
del indicators_df['Series Code']
del indicators_df['Unnamed: 6']
del indicators_df['Unnamed: 7']
del indicators_df['Data Source']
del indicators_df['Unnamed: 9']
del indicators_df['2020 [YR2020]']
```

```python
indicators_df.head()
```

```python
#make lists with the information we need
countries_name = indicators_df['Country Code'].unique()
series_name = indicators_df['Series Name'].unique()
year = indicators_df['2018 [YR2018]'].unique()
```

```python
#create a new dataframe from these lists
final_df = pd.DataFrame({'Country Code': indicators_df['Country Code'],
                   'Series Name': indicators_df['Series Name'],
                   '2018 [YR2018]': indicators_df['2018 [YR2018]']})
final_df = final_df. dropna() #drop the NAN which show missing data as it creates a bug when doing pivot()
final_df.head()
```

```python
#pivot the dataframe from long to wide
final_df = pd.pivot(final_df, index='Country Code', columns='Series Name', values='2018 [YR2018]')
```
